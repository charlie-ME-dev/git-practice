## Lab 3-5: if-elif-else 문을 사용한 다중 조건식

### 1. 자동차의 속도(speed)가 100 이상이면 '고속', 60 이상 100 미만이면 '중속', 60 미만이면 '저속'을 출력하는 프로그램을 작성하여라.

```python
speed = int(input("자동차의 속도를 입력하세요(단위 : km/h)")) 
# 마찬가지로 실수입력하면 ValueError 등장 else이므로 음수넣으면 저속이라고 한다.
if speed >= 100:
    print('고속')
elif speed >= 60:
    print('중속')
else:
    print('저속')
```

## Lab 4-3: 누적 덧셈의 응용

### 1~3. 누적 덧셈을 응용하여 다양한 정수의 합 구하기

```python
# 1. 1에서 100까지 정수의 합
s1 = 0
for i in range(1, 101):
    s1 += i
print('1에서 100까지의 합:', s1)

# 2. 1에서 100까지 정수 중에서 짝수의 합
s2 = 0
for i in range(0, 101, 2):
    s2 += i
print('1에서 100까지 짝수의 합:', s2)

# 3. 1에서 100까지 정수 중에서 홀수의 합
s3 = 0
for i in range(1, 101, 2):
    s3 += i
print('1에서 100까지 홀수의 합:', s3)
```
