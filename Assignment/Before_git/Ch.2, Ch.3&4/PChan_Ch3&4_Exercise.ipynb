### [2] 4.4 맛나 식당의 메뉴 주문 프로그램을 개발하고자 한다. 이를 위하여 사용자에게 다음과 같은 1), 2), 3)의 메뉴를 보여주고 이 중에서 하나를 선택하도록 하자. 이때 다음 보기와 같이 메뉴에 없는 5, 6, 0과 같은 숫자가 입력되면 '메뉴를 다시 입력하세요: '가 출력되도록 한 후 다시 입력을 받도록 하자

```python
print('맛나 식당에 오신 것을 환영합니다. 메뉴는 다음과 같습니다.')
print('1) 햄버거')
print('2) 치킨')
print('3) 피자')

while True:
    menu = input('1에서 3까지의 메뉴를 선택하세요 : ')
    menu = int(menu)
    if 1<= menu <=3:
        break
    else:
        print('메뉴를 다시 입력하세요:')

if menu == 1:
    print('햄버거를 선택하였습니다.')
elif menu == 2:
    print('치킨을 선택하였습니다.')
else :
    print('피자를 선택하였습니다.')
```

### [3] 4.5 맛나 식당의 메뉴 주문 프로그램을 개발하고자 한다. 이를 위하여 사용자에게 다음과 같은 메뉴를 보여주고 이 중에서 하나의 알파벳 b, c, p를 선택하도록 하자. 이때 메뉴에 없는 알파벳이 입력되면 '메뉴를 다시 입력하세요: '가 출력되도록 한 후 다시 입력을 받도록 하자.

```python
print('맛나 식당에 오신 것을 환영합니다. 메뉴는 다음과 같습니다.')
print('- 햄버거(입력 b)')
print('- 치킨(입력 c)')
print('- 피자(입력 p)')

while True:
    menu = input('메뉴를 선택하세요(알파벳(b, c, p 입력): )')
    menu = str(menu)
    if menu == 'b' or menu == 'c' or menu == 'p':
        break
    else:
        print('메뉴를 다시 입력하세요(알파벳(b, c, p 입력): ')

if menu == 'b':
    print('햄버거를 선택하였습니다.')
elif menu == 'c':
    print('치킨을 선택하였습니다.')
else:
    print('피자를 선택하였습니다.')

# 문자열은 '' 하는거 까먹지 말자.
```

### [4] 4.9 사용자로부터 양의 정수 n을 입력 받은 후 $1^2 + 2^2 + 3^2 + \dots + n^2$ 을 구하는 프로그램을 작성하시오.

```python
while True:
    number = input('숫자를 입력하세요 : ')
    number = int(number)
    if number > 0:
        break
    else:
        print('양의 정수를 입력하세요')

output = 0
# initialization이게 중요하네.. 처음엔 number에 더해버리니까 이상해지더라..
for i in range(1,number+1):
    output += i**2
print('결과는', output, '입니다. ')
```
