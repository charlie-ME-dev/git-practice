## [1] 5.3 
m과 n이라는 이름의 매개변수를 두 개 입력 받아서 이 두 값 중에서 큰 값을 반환하는 `max2(m,n)` 함수와 작은 값을 반환하는 `min2(m,n)` 함수를 구현하시오. 100과 200을 인자로 넣어서 호출하여 그 결과를 확인하시오

```python
def max2(m,n):
    upper_value = max(m,n)
    return upper_value

def min2(m,n):
    lower_value = min(m,n)
    return lower_value

print('100과 200중 큰 수는:',max2(100,200))
print('100과 200중 작은 수는:',min2(100,200))
```

## [2] 5.5
인치 값을 매개변수로 넘겨받아서 이를 센티미터(cm) 값으로 반환하는 `inch2cm(inch)` 함수를 구현하고 이 함수를 호출하여 1에서 5인치를 센티미터로 출력하시오(단, 1인치는 2.54 센티미터로 정의하시오.)

```python
def inch2cm(inch):
    centimeter = inch*2.54
    return centimeter


for i in range(1,6):
    print(f'{i}인치는 {inch2cm(i):.2f}센티미터')
```

## [3] 5.9
사용자로부터 임의의 수를 연속적으로 입력받도록 하시오. 이 수들에 대한 평균값, 최댓값, 최솟값을 반환하는 함수 `mean_of_n(nums)`, `min_of_n(nums)`를 구현하여 다음과 같이 출력하여라.

정수를 여러 개 입력하시오 : 3 45 32 5 7 8 4 44 5 90 17  
평균값은 23.6  
최댓값은 90  
최솟값은 3

```python
def mean_of_n(nums):
    mean = sum(nums)/len(nums)
    return mean
def max_of_n(nums):
    maximum = max(nums)
    return maximum
def min_of_n(nums):
    minimum = min(nums)
    return minimum

raw = input('정수를 여러 개 입력하시오 : ')
numnum = [int(x) for x in raw.split()]

print(f'평균값은 {mean_of_n(numnum):.1f}')
print(f'최댓값은 {max_of_n(numnum)}')
print(f'최솟값은 {min_of_n(numnum)}')
```

## [4] 5.17
n1에서 n2까지의 정수의 합을 구하는 함수 `sum_range(n1,n2)`를 작성하시오. 이 함수를 이용하여 10에서 20까지의 정수의 합과 40에서 100까지의 정수의 합을 구하여 다음과 같이 출력하는 프로그램을 작성하시오. 이때 모든 숫자 값 출력문에는 플레이스 홀더를 사용하시오.

[실행결과]

10에서 20까지의 정수의 합 : 165

40에서 100까지의 정수의 합 : 4270

```python
def sum_range(n1,n2):
    summation = 0
    for i in range(n1,n2+1,1):
        summation += i
    return summation
print('10에서 20까지의 정수의 합: {}'.format(sum_range(10,20)))
print('40에서 100까지의 정수의 합: {}'.format(sum_range(40,100)))
```
