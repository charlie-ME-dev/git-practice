## Lab 5-6: 함수의 사용

### 1. 다음 2차 방정식의 근을 [코드 5-10]의 함수를 사용하여 출력하여라.

```python
def print_root(a, b, c):
    r1 = (-b + (b**2 - 4*a*c)**0.5) / (2*a)
    r2 = (-b - (b**2 - 4*a*c)**0.5) / (2*a)
    print('해는', r1, '또는', r2)

print("(1) x^2 + 4x - 21 = 0")
print_root(1, 4, -21)

print("(2) x^2 - 6x + 8 = 0")
print_root(1, -6, 8)
```

## Lab 5-7: 원의 면적과 둘레를 반환하는 함수

### 1. 반지름을 받아 면적과 둘레를 반환하는 함수

```python
def circle_area_circum(radius):
    area = 3.14 * radius ** 2
    circum = 2 * 3.14 * radius
    return area, circum 

radius = 10
a, c = circle_area_circum(radius)
print(f"반지름 {radius}인 원의 면적은 {a:.1f}, 원의 둘레는 {c:.1f}")
```

## Lab 5-8: 다수의 결과를 반환하는 함수 만들기
### `n`의 배수 `m`개를 반환하는 `multiples(n, m)`

```python
def multiples(n, m):
    res = []
    for i in range(1, m + 1):
        res.append(n * i)
    return tuple(res) 


r1, r2, r3, r4 = multiples(3, 4)
print(r1, r2, r3, r4)

r1,r2,r3,r4,r5 = multiples(2,5)
print(r1,r2,r3,r4,r5)
```

# Lab 5-10: 가변 인자의 활용

### 5.1 가변 인자를 사용하는 `sum_nums()` 함수를 수정하여 인자들을 튜플 형식으로 출력한 후 모든 값들의 합과 평균을 다음과 같이 출력하시오.

```python
def sum_num(*numbers):
    result = 0
    for n in numbers:
        result += n
    mean = result / len(numbers)
    print(f'{len(numbers)} 개의 인자 {numbers}')
    print(f'합계 : {result}, 평균 : {mean:.1f}')

sum_num(10,20,30)
sum_num(10,20,30,40,50)
```

### 5.2 가변 인자를 사용하는 `min_nums()` 함수를 구현하시오. 이 함수는 정수를 인자로 받을 수 잇는데 이 인자의 개수가 가변적이다. 이 함수의 호출문이 다음과 같을 경우 `min_nums(20,40,50,10)` 다음과 같은 출력이 나타나도록 하시오. `최솟값은 10`

```python
def min_nums(*numbers):
    low = min(numbers)
    print(f'최솟값은 {low}')

min_nums(20,40,50,10)
```
