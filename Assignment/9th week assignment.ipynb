### 1. datetime 모듈을 사용하여 오늘의 날짜와 시간을 다음과 같이 출력하여라. 이때 hour 값이 10일 경우 오전 10시, 13이면 오후 1시와 같이 오전 오후 정보를 출력하여라.
오늘의 날짜 : 2025년 2월 19일
현재시간 : 오후 6시 50분 11초

```python
from datetime import datetime as dt

start_time = dt.now()

year = start_time.year
month = start_time.month
day = start_time.day

hour = start_time.hour
minute = start_time.minute
second = start_time.second

if hour < 12:
    am_pm = '오전'
    display_hour = hour
else:
    am_pm = '오후'
    display_hour = hour -12

if display_hour == 0:
    display_hour = 12

print(f'오늘의 날짜: {year}년 {month}월 {day}일')
print(f'현재시간 : {am_pm} {display_hour}시 {minute}분 {second}초')
```

## 8-2 1부터 1,000,000까지 정수의 합을 구하여 반환하는 함수 sun1to1000000()을 작성하여라.
### (1) 이 계산을 수행하는 데 걸리는 시간을 다음과 같이 초 단위로 구하여 출력하여라.

1에서 1,000,000까지의 합을 구하는 시간 : 0.0573초

```python
import time

def sum1to1000000():
    total = 0

    for i in range(1, 1000001):
        total += i

    return total


start_time = time.time()

result = sum1to1000000()

end_time = time.time()

elapsed_time = end_time - start_time

print(f"1에서 1,000,000까지의 합을 구하는 시간 : {elapsed_time:.4f}초")
```

### (2) 10000 팩토리얼을 구하는 함수

10000!을 100번 반복해서 궇나느 시간 : 0.226초

```python
import time

def factorial_10000():
    result = 1

    for i in range(1, 10001):
        result *= i

    return result


start_time = time.time()

for i in range(100):
    result = factorial_10000()

end_time = time.time()

elapsed_time = end_time - start_time

print(f"10000!을 100번 반복해서 구하는 시간 : {elapsed_time:.4f}초")
```

### (4) 1도에서 360도 사이의 sin값을 1도 단위로 구하여 그 합을 반환하는 함수

1에서 360도까지 sin 값의 합을 100번 반복하는 시간 : 0.0521초

```python
import time
import math

def sum_sin_values():
    total = 0

    for degree in range(1, 361):
        radian = math.radians(degree)
        total += math.sin(radian)

    return total


start_time = time.time()

for i in range(100):
    result = sum_sin_values()

end_time = time.time()

elapsed_time = end_time - start_time

print(f"1에서 360도까지 sin 값의 합을 100번 반복하는 시간 : {elapsed_time:.4f}초")
```

## 8.4 0부터 1,000,000까지의 임의의 정수를 다음과 같이 연속적으로 10개 출력하는 의사난수 함수 my_rand()를 생성하여라.
### (1) 이 의사난수를 만들기 위한 계수 값과 식을 적으시오.

434737

338960

(...중간생략...)

474664

의사난수 생성식:
X_{n+1} = (1103515245 * X_n + 12345) % 2^31

출력값:
X_{n+1} % 1000001

```python
seed = 123456

def my_rand():
    global seed

    a = 1103515245
    c = 12345
    m = 2 ** 31

    seed = (a * seed + c) % m

    return seed % 1000001


for i in range(10):
    print(my_rand())
```

### (2) 이 의사난수를 만들기 위한 초기 시드 값으로 time 모듈의 time() 함수를 사용하여 다시 코딩하시오. (힌트: time() 값은 실수 값이므로 이를 정수로 변환하여 사용해야 한다). 이번에는 0에서 1,000,000까지의 임의의 5개의 정수를 출력하고 이를 다음과 같이 2번 반복하여라.

```python
import time

seed = int(time.time() * 1000)

def my_rand():
    global seed

    a = 1103515245
    c = 12345
    m = 2 ** 31

    seed = (a * seed + c) % m

    return seed % 1000001

for i in range(2):
    for j in range(5):
        print(my_rand())
    print()
```
