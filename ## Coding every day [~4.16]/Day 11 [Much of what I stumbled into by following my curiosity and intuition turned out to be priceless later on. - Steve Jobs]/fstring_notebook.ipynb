## 1. f-string 소개 / Introduction

**f-string**은 Python 3.6부터 도입된 문자열 포맷팅 방법입니다.

### 기존 방법과의 비교 / Comparison with Old Methods

```python
# 변수 정의 / Define variables
name = "김철수"
age = 25

# 1. % 포맷팅 (구식) / Old % formatting
msg1 = "이름: %s, 나이: %d" % (name, age)
print("방법 1:", msg1)

# 2. .format() 메서드 / .format() method
msg2 = "이름: {}, 나이: {}".format(name, age)
print("방법 2:", msg2)

# 3. f-string (권장) / f-string (recommended) ⭐
msg3 = f"이름: {name}, 나이: {age}"
print("방법 3:", msg3)
```

## 3. 표현식 사용 / Using Expressions

### 3.1 산술 연산 / Arithmetic Operations

```python
# 직사각형 넓이 계산 / Rectangle area calculation
width = 10
length = 15

print(f"가로: {width}cm, 세로: {length}cm")
print(f"넓이: {width * length}cm²")
print(f"둘레: {2 * (width + length)}cm")
```

```python
# 평균 계산 / Average calculation
korean = 85
english = 92
math = 88

average = f"평균 점수: {(korean + english + math) / 3:.2f}점"
print(average)
```

### 3.2 함수 호출 / Function Calls

```python
# 문자열 메서드 사용 / Using string methods
text = "hello world"

print(f"원본: {text}")
print(f"대문자: {text.upper()}")
print(f"첫 글자 대문자: {text.title()}")
print(f"길이: {len(text)}자")
```

```python
# 리스트 함수 / List functions
scores = [85, 92, 78, 95, 88]

print(f"점수 목록: {scores}")
print(f"최고 점수: {max(scores)}점")
print(f"최저 점수: {min(scores)}점")
print(f"평균: {sum(scores) / len(scores):.1f}점")
```

### 3.3 조건 표현식 / Conditional Expressions

```python
# 합격/불합격 판정 / Pass/Fail evaluation
score = 75
result = f"점수: {score}점 - {'합격' if score >= 60 else '불합격'}"
print(result)
```

```python
# 학점 계산 / Grade calculation
score = 87
grade = 'A' if score >= 90 else 'B' if score >= 80 else 'C' if score >= 70 else 'D' if score >= 60 else 'F'
print(f"{score}점 → 학점: {grade}")
```

## 4. 숫자 포맷팅 / Number Formatting

### 4.1 소수점 자리수 / Decimal Places

```python
# 원주율 표현 / Pi representation
pi = 3.141592653589793

print(f"원본: {pi}")
print(f"소수점 2자리: {pi:.2f}")
print(f"소수점 4자리: {pi:.4f}")
print(f"소수점 6자리: {pi:.6f}")
```

```python
# 금액 계산 / Price calculation
price = 15000
quantity = 3
discount_rate = 0.15

total = price * quantity
discount = total * discount_rate
final_price = total - discount

print(f"단가: {price}원")
print(f"수량: {quantity}개")
print(f"소계: {total}원")
print(f"할인({discount_rate:.0%}): -{discount:.0f}원")
print(f"최종 금액: {final_price:.0f}원")
```

### 4.2 천단위 구분자 / Thousands Separator

```python
# 큰 숫자 표현 / Large number representation
population = 51000000
gdp = 1700000000000

print(f"인구: {population:,}명")
print(f"GDP: {gdp:,}원")
print(f"1인당 GDP: {gdp/population:,.0f}원")
```

### 4.3 퍼센트 표현 / Percentage

```python
# 비율을 퍼센트로 / Ratio to percentage
correct = 85
total = 100
accuracy = correct / total

print(f"정답: {correct}/{total}")
print(f"정확도: {accuracy:.1%}")
print(f"정확도: {accuracy:.2%}")
```

### 4.4 진법 변환 / Base Conversion

```python
# 다양한 진법으로 표현 / Various base representations
num = 255

print(f"10진수 (Decimal): {num}")
print(f"2진수 (Binary): {num:b}")
print(f"8진수 (Octal): {num:o}")
print(f"16진수 (Hexadecimal): {num:x}")
print(f"16진수 대문자 (Hex Upper): {num:X}")
```

## 5. 문자열 정렬 / String Alignment

### 5.1 기본 정렬 / Basic Alignment

```python
# 왼쪽/가운데/오른쪽 정렬 / Left/Center/Right alignment
word = "Python"

print(f"왼쪽 정렬 (Left):    '{word:<20}'")
print(f"가운데 정렬 (Center): '{word:^20}'")
print(f"오른쪽 정렬 (Right):  '{word:>20}'")
```

### 5.2 패딩 문자 지정 / Padding Characters

```python
# 다양한 패딩 문자 / Various padding characters
title = "Menu"

print(f"{title:*<30}")
print(f"{title:=^30}")
print(f"{title:->30}")
```

### 5.3 표 만들기 / Creating Tables

```python
# 제품 목록 표 / Product list table
print("="*50)
print(f"{'제품명':<15} {'가격':>15} {'재고':>10}")
print("="*50)
print(f"{'노트북':<15} {1500000:>15,} {25:>10}")
print(f"{'마우스':<15} {25000:>15,} {150:>10}")
print(f"{'키보드':<15} {85000:>15,} {80:>10}")
print(f"{'모니터':<15} {350000:>15,} {45:>10}")
print("="*50)
```

## 6. 날짜/시간 포맷팅 / DateTime Formatting

```python
from datetime import datetime

now = datetime.now()

# 다양한 날짜 형식 / Various date formats
print(f"전체: {now}")
print(f"날짜: {now:%Y-%m-%d}")
print(f"시간: {now:%H:%M:%S}")
print(f"한국식: {now:%Y년 %m월 %d일}")
print(f"시간 포함: {now:%Y년 %m월 %d일 %H시 %M분}")
```

```python
# 날짜 계산 / Date calculation
from datetime import datetime, timedelta

today = datetime.now()
tomorrow = today + timedelta(days=1)
next_week = today + timedelta(weeks=1)

print(f"오늘: {today:%Y-%m-%d}")
print(f"내일: {tomorrow:%Y-%m-%d}")
print(f"다음 주: {next_week:%Y-%m-%d}")
```

## 7. 고급 기능 / Advanced Features

### 7.1 딕셔너리 사용 / Using Dictionaries

```python
# 학생 정보 딕셔너리 / Student information dictionary
student = {
    "name": "박민수",
    "id": "2024001",
    "grade": 3,
    "major": "컴퓨터공학",
    "gpa": 3.85
}

profile = f"""
학생 프로필 / Student Profile
{'='*40}
이름: {student['name']}
학번: {student['id']}
학년: {student['grade']}학년
전공: {student['major']}
평점: {student['gpa']:.2f}
{'='*40}
"""
print(profile)
```

### 7.2 디버깅 기능 (Python 3.8+) / Debugging (Python 3.8+)

```python
# = 연산자로 변수명과 값 함께 출력 / Print variable name and value with =
x = 42
y = 3.14
name = "Python"

print(f"{x=}")
print(f"{y=:.2f}")
print(f"{name=}")
print(f"{x + y=}")
```

### 7.3 멀티라인 f-string / Multi-line f-string

```python
# 보고서 생성 / Generate report
university = "경북대학교"
department = "컴퓨터학부"
students = 500
professors = 25
year = 2024

report = f"""
{'='*50}
       대학 정보 보고서 / University Report
{'='*50}
학교명: {university}
학과명: {department}
학생 수: {students:,}명
교수 수: {professors}명
학생/교수 비율: {students/professors:.1f}:1
보고 연도: {year}년
{'='*50}
"""
print(report)
```

## 9. 정리 / Summary

### f-string 주요 문법 정리 / Key Syntax Summary

| 기능 / Feature | 문법 / Syntax | 예시 / Example |
|---------------|--------------|---------------|
| 기본 변수 / Basic | `f"{var}"` | `f"{name}"` |
| 소수점 / Decimal | `f"{var:.nf}"` | `f"{pi:.2f}"` |
| 천단위 구분 / Comma | `f"{var:,}"` | `f"{price:,}"` |
| 퍼센트 / Percent | `f"{var:.n%}"` | `f"{ratio:.1%}"` |
| 왼쪽 정렬 / Left | `f"{var:<n}"` | `f"{text:<10}"` |
| 가운데 정렬 / Center | `f"{var:^n}"` | `f"{text:^10}"` |
| 오른쪽 정렬 / Right | `f"{var:>n}"` | `f"{text:>10}"` |
| 2진수 / Binary | `f"{var:b}"` | `f"{num:b}"` |
| 16진수 / Hex | `f"{var:x}"` | `f"{num:x}"` |
| 디버깅 / Debug | `f"{var=}"` | `f"{x=}"` |
