## 10. 대칭수(palindrome)는 앞에서부터 읽으나 뒤에서부터 읽으나 그 값이 같은 수를 말합니다. 두 자리 수 두 개를 곱해 만들 수 있는 가장 큰 대칭수는 9009 = 91 * 99 입니다. 세 자리 수 두 개를 곱해 만들 수 있는 가장 큰 대칭수를 구하세요. 결과를 대칭수 : [값], 인수 : [값1], [값2] 형식으로 출력하시오. 최적화된 루프를 사용하여 실행 시간을 단축해 보시오. 이 경우에는 가점을 드립니다.

```python
def palindrome_three_digit():
    pal_number = 0
    first_val = 0
    second_val = 0
    
    for i in range(100,1000):
        for j in range(i,1000):
            cal_number = str(i*j)
            if cal_number == cal_number[::-1]:
                pal_number = cal_number
                first_val = i
                second_val = j
                
    
    print(f'대칭수 : [{pal_number}], 인수: [{first_val}],[{second_val}]')


palindrome_three_digit() # 888888 924 962
```
