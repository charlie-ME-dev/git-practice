```python
def calculate_display(diagonal_inch, aspect_ratio = '16:9'):
    diagonal_cm = diagonal_inch * 2.54

    if diagonal_inch > 100 or diagonal_inch < 10:
        print('Invalid screen Size')
        return None
    supported_ratio = ['16:9', '16:10','21:9','4:3']

    if aspect_ratio not in supported_ratio:
        print('Invalid aspect ratio')
        return None

    W_str, h_str = aspect_ratio.split(':')
    width = float(W_str)
    height = float(h_str)

    width_cm = diagonal_cm * (width) / (width**2 + height**2)**0.5
    height_cm = diagonal_cm * (height) / (width**2 + height**2)**0.5
    
    area_cm2 = width_cm * height_cm
    viewing_distance_cm = diagonal_cm * 1.5

    return round(diagonal_cm,2), round(width_cm,2), round(height_cm,2), round(area_cm2,2), round(viewing_distance_cm,2)


# 55-inch 16:9 TV
result = calculate_display(55)
print(result)
# (139.7, 121.76, 68.49, 8340.19, 209.55)

diagonal, width, height, area, distance = calculate_display(55)
print(f"Width: {width}cm, Height: {height}cm")
# Width: 121.76cm, Height: 68.49cm

# 27-inch 16:9 monitor
result = calculate_display(27, '16:9')
print(result)
# (68.58, 59.77, 33.62, 2009.55, 102.87)

# 34-inch 21:9 ultrawide monitor
result = calculate_display(34, '21:9')
print(result)
# (86.36, 79.89, 34.24, 2735.52, 129.54)

# 24-inch 16:10 monitor
result = calculate_display(24, '16:10')
print(result)
# (60.96, 51.69, 32.31, 1670.1, 91.44)

# Error cases
result = calculate_display(5)          # Invalid screen size, returns None
result = calculate_display(55, '32:9') # Invalid aspect ratio, returns None

    
```
