## 7.1 딕셔너리 자료형 / Dictionary Data Type

**딕셔너리(dictionary)** 는 **키(key)** 와 **값(value)** 을 쌍으로 가지는 자료형입니다.  
A **dictionary** is a data type that stores data as **key-value pairs**.

- 키를 이용하여 값을 참조함 / Values are referenced by their keys
- 키는 중복될 수 없음 / Keys must be unique
- 문법 / Syntax: `{ key1: value1, key2: value2, ... }`

```python
# 딕셔너리 생성 (Creating a dictionary)
# 항목은 [키]:[값] 형식으로 콜론으로 연결됨
# Items are written as key:value pairs separated by colons
person = {'name': '홍길동', 'age': 25, 'job': 'student'}
print(person)

# 키를 이용하여 값 참조 (Access value by key)
print("name:", person['name'])
print("age :", person['age'])
```

```python
# 리스트와의 비교 — 리스트는 0부터 시작하는 인덱스로 접근
# Contrast with list — lists use 0-based integer indexes
fruits = ['banana', 'apple', 'orange', 'kiwi']  # 문자열을 가지는 리스트 / list of strings
print(fruits[0])   # 인덱스 0 → 'banana'
print(fruits[2])   # 인덱스 2 → 'orange'

# 3명의 학생에 대한 학번: 이름 딕셔너리 생성
# Create a dictionary mapping student IDs to names
students = {20230001: '김철수', 20230002: '이영희', 20230003: '박민수'}
print(students)
print("학번 20230002의 이름:", students[20230002])
```

## 7.2 딕셔너리의 삽입과 삭제 / Dictionary Insertion and Deletion

- 삽입 / Insertion: `dict[new_key] = new_value`
- 삭제 / Deletion: `del dict[key]`
- 존재하지 않는 키를 삭제하면 `KeyError` 발생  
  Deleting a non-existent key raises `KeyError`

```python
# 새 항목 삽입 (Insert a new item)
person = {'name': '홍길동', 'age': 25}
person['job'] = 'engineer'    # 'job' 키가 없으면 새로 추가됨 / adds a new key
person['age'] = 26            # 'age' 키가 이미 있으면 값이 갱신됨 / updates existing
print("삽입/수정 후:", person)

# 항목 삭제 (Delete an item)
del person['job']
print("삭제 후:", person)
```

```python
# 존재하지 않는 키를 삭제하면 KeyError가 발생함
# Deleting a non-existent key raises KeyError
try:
    del person['hobby']   # 'hobby' 키는 존재하지 않음 / 'hobby' does not exist
except KeyError as e:
    print(f"KeyError 발생 / KeyError raised: {e}")
```

## 7.3 딕셔너리와 연산자 / Dictionaries and Operators

| 연산자 / Operator | 설명 / Description |
|---|---|
| `len()` | 항목 수 반환 / Number of items |
| `in`, `not in` | 키 존재 여부 확인 / Check key existence |
| `==`, `!=` | 두 딕셔너리의 동일성 비교 / Equality check |

⚠️ 딕셔너리는 `<`, `<=`, `>`, `>=` 비교를 지원하지 **않음**.  
⚠️ Dictionaries do **not** support `<`, `<=`, `>`, `>=`.

```python
person = {'name': '홍길동', 'age': 25, 'job': 'student'}

# len(): 항목의 수 / number of key-value pairs
print("항목 수 / number of items:", len(person))

# in / not in: 키가 딕셔너리에 있는지 확인 / key existence check
print("'name' in person  :", 'name' in person)     # True
print("'hobby' in person :", 'hobby' in person)    # False
print("'hobby' not in person :", 'hobby' not in person)  # True

# == / !=: 동일성 비교 / equality comparison
d1 = {'a': 1, 'b': 2}
d2 = {'b': 2, 'a': 1}  # 순서는 달라도 동일함 / order-independent
d3 = {'a': 1, 'b': 3}
print("d1 == d2 :", d1 == d2)   # True
print("d1 == d3 :", d1 == d3)   # False
```

```python
# 비교 연산자는 지원하지 않음 → TypeError
# Comparison operators are NOT supported → TypeError
try:
    print(d1 < d3)
except TypeError as e:
    print(f"TypeError 발생 / TypeError raised: {e}")
```

## 7.4 딕셔너리의 메소드 / Dictionary Methods

| 메소드 / Method | 반환값 / Returns |
|---|---|
| `.keys()`   | 모든 키 / all keys |
| `.values()` | 모든 값 / all values |
| `.items()`  | (키, 값) 튜플 / (key, value) tuples |
| `.get(k)`   | 키가 있으면 값, 없으면 `None` / value if key exists, else `None` |

```python
person = {'이름': '홍길동', '나이': 25, '직업': '학생'}

# .keys() — 키들의 뷰 객체 반환 / a view of keys
print("keys  :", person.keys())

# .values() — 값들의 뷰 객체 반환 / a view of values
print("values:", person.values())

# .items() — (키, 값) 튜플들의 뷰 반환 / a view of (key, value) tuples
# 각 항목은 튜플(tuple) 형태로 반환됨 / each item is a tuple like ('이름', '홍길동')
print("items :", person.items())
```

```python
# 존재하지 않는 키를 대괄호로 접근하면 KeyError 발생
# Accessing a missing key with [] raises KeyError
try:
    print(person['취미'])
except KeyError as e:
    print(f"KeyError 발생 / KeyError raised: {e}")

# .get()은 키가 없어도 예외 없이 None을 반환 — 안전한 조회 방법
# .get() returns None (or a default) instead of raising — safer lookup
print("person.get('취미')        :", person.get('취미'))              # None
print("person.get('취미', '없음') :", person.get('취미', '없음'))      # 기본값 사용 / default
print("person.get('이름')        :", person.get('이름'))              # '홍길동'
```

```python
# 딕셔너리의 for-in 반복은 기본적으로 '키(key)'를 반환함
# Iterating a dict with for-in yields the KEYS (not values)
print("--- for k in person  (키만 반환) ---")
for k in person:
    print(k, '->', person[k])

# 키와 값을 동시에 얻고 싶다면 .items() 사용
# To iterate over keys and values together, use .items()
print("--- for k, v in person.items() ---")
for k, v in person.items():
    print(f"{k}: {v}")
```

## 7.5 리스트와 딕셔너리의 비교 / List vs Dictionary

| 항목 / Aspect | 리스트 / List | 딕셔너리 / Dictionary |
|---|---|---|
| 생성 / Create | `[v1, v2, ...]` | `{k1:v1, k2:v2, ...}` |
| 접근 / Access | `a[index]` (정수 인덱스) | `d[key]` (임의 키) |
| 길이 / Length | `len(a)` | `len(d)` |
| 삭제 / Delete | `del a[index]` | `del d[key]` |
| 포함 / Contains | `value in a` | `key in d` |

```python
# 생성과 접근 비교 / Creation and access side-by-side
fruits_list = ['banana', 'apple', 'orange']
fruits_dict = {'b': 'banana', 'a': 'apple', 'o': 'orange'}

print("list [0] ->", fruits_list[0])   # 정수 인덱스로 접근 / integer index
print("dict['a']->", fruits_dict['a']) # 키로 접근 / key lookup

# 항목 수, 포함 여부, 삭제
# length, membership, deletion
print("len(list) =", len(fruits_list), ", len(dict) =", len(fruits_dict))
print("'apple' in list :", 'apple' in fruits_list)   # 리스트: 값 검색 / list: searches VALUES
print("'a' in dict     :", 'a' in fruits_dict)       # 딕셔너리: 키 검색 / dict: searches KEYS

del fruits_list[0]
del fruits_dict['b']
print("after delete — list:", fruits_list, "| dict:", fruits_dict)
```

## 7.6 튜플 자료형 / Tuple Data Type

**튜플(tuple)** 은 여러 요소를 가진 **불변(immutable)** 컬렉션입니다.  
A **tuple** is an **immutable** collection of multiple elements.

- 문법 / Syntax: `t = (v1, v2, v3)` — 괄호로 둘러쌈 / wrapped in parentheses
- 한 번 정해진 요소의 순서를 바꿀 수 없음 / order cannot be changed
- 내부 객체를 변경/삭제할 수 없음 / items cannot be changed or deleted
- 인덱싱은 리스트와 동일 / indexing works like lists

```python
# 튜플 생성의 여러 가지 방법 / Several ways to create tuples
t1 = ('one', 'two', 'three')     # 괄호 사용 / with parentheses
t2 = 'one', 'two', 'three'       # 괄호 생략 가능 / parentheses optional
t3 = tuple(['one', 'two'])       # 리스트로부터 변환 / from a list
t4 = ()                          # 빈 튜플 / empty tuple
t5 = (42,)                       # 원소가 1개인 튜플 — 반드시 콤마 / single-element needs trailing comma
t_not_tuple = (42)               # 콤마가 없으면 그냥 정수 42 / without comma this is just int 42

print("t1:", t1, type(t1))
print("t2:", t2, type(t2))
print("t3:", t3, type(t3))
print("t4:", t4, type(t4))
print("t5:", t5, type(t5))
print("t_not_tuple:", t_not_tuple, type(t_not_tuple))
```

```python
# 인덱스로 접근은 리스트와 동일 / Indexing works like lists
t = (10, 20, 30)
print("t[0] =", t[0])
print("t[-1]=", t[-1])

# 그러나 개별 항목 수정은 불가능 → TypeError
# But assigning to an element raises TypeError (immutability)
try:
    t[0] = 999
except TypeError as e:
    print(f"TypeError 발생 / TypeError raised: {e}")
```

### 패킹과 언패킹 / Packing and Unpacking

- **패킹(packing)**: 여러 값을 하나의 튜플로 묶는 것 / putting several values into a tuple
- **언패킹(unpacking)**: 튜플의 각 값을 개별 변수에 꺼내는 것 / extracting values into separate variables

```python
# 패킹(packing) — 여러 값을 하나의 튜플로
# Packing — multiple values into one tuple
packed = 1, 2, 3
print("packed:", packed)

# 언패킹(unpacking) — 튜플의 값을 각 변수에 분배
# Unpacking — distribute values into separate variables
a, b, c = packed
print(f"a={a}, b={b}, c={c}")

# 변수 개수가 맞지 않으면 ValueError 발생
# Wrong number of variables → ValueError
try:
    x, y = packed
except ValueError as e:
    print(f"ValueError 발생 / ValueError raised: {e}")
```

```python
# 스왑(swap) — 튜플을 이용한 가장 직관적인 값 교환
# Swap — the Pythonic idiom for swapping two variables
a, b = 10, 20
print(f"교환 전 / before: a={a}, b={b}")

# 오른쪽에서 튜플 (b, a)가 만들어지고 왼쪽에서 언패킹됨
# On the right, the tuple (b, a) is created; on the left it is unpacked
a, b = b, a
print(f"교환 후 / after : a={a}, b={b}")
```

## 7.7 튜플의 연산 / Tuple Operations

- `+` : 튜플끼리 연결 / concatenation (tuple + tuple)
- `*` : 튜플 반복 / repetition (tuple * int)
- 튜플 + 정수는 **오류** — 튜플 + 튜플이어야 함  
  tuple + int is an **error** — both operands must be tuples
- 튜플의 값을 바꾸고 싶다면 `list()` → 수정 → `tuple()` 과정을 거쳐야 함  
  To "change" a tuple, convert to list, edit, convert back

```python
t0 = (10, 20, 30)

# + 연산: 튜플 + 튜플 연결 / concatenation
print("t0 + (40,)   :", t0 + (40,))        # (10, 20, 30, 40)
print("t0 + (40, 50):", t0 + (40, 50))

# * 연산: 반복 / repetition
print("t0 * 2 :", t0 * 2)                  # (10, 20, 30, 10, 20, 30)

# 튜플 + 정수는 불가 — TypeError
# tuple + int is NOT allowed — TypeError
try:
    print(t0 + 40)
except TypeError as e:
    print(f"TypeError 발생 / TypeError raised: {e}")
```

```python
# 튜플의 값을 '바꾸기' — list()와 tuple() 변환을 사용
# 'Modifying' a tuple — go through list and back
t = (1, 2, 3, 4)
print("원본 / original:", t)

temp = list(t)       # 1) 리스트로 변환 / convert to list
temp[1] = 99         # 2) 리스트에서 수정 / modify the mutable list
t = tuple(temp)      # 3) 다시 튜플로 변환 / convert back to tuple

print("수정 후 / modified:", t)
```

## 7.8 함수 반환문과 튜플 활용 / Tuples as Function Return Values

파이썬 함수는 여러 값을 **한 번에 반환**할 수 있습니다. 내부적으로는 튜플로 묶여 반환됩니다.  
A Python function can return multiple values at once — they are packed into a tuple.

```python
# 몫과 나머지를 함께 반환하는 함수
# A function that returns both quotient and remainder
def divmod_custom(a, b):
    """a를 b로 나눈 몫과 나머지를 튜플로 반환 / return (quotient, remainder) as a tuple."""
    return a // b, a % b       # 실제로는 (a // b, a % b) 튜플을 반환 / returns a tuple

# 반환값을 튜플 하나로 받을 수도 있고…
# The caller can receive it as one tuple…
result = divmod_custom(17, 5)
print("tuple 하나로 / as a single tuple:", result, type(result))

# …언패킹을 통해 여러 변수로 나눠 받을 수도 있음
# …or unpack it into separate variables
q, r = divmod_custom(17, 5)
print(f"몫 / quotient = {q}, 나머지 / remainder = {r}")
```

```python
# 최소값과 최대값을 함께 반환하는 예제
# Example: return both min and max at once
def min_max(numbers):
    return min(numbers), max(numbers)   # 튜플로 묶여 반환 / packed into a tuple

lo, hi = min_max([3, 1, 4, 1, 5, 9, 2, 6])
print(f"최소 / min = {lo}, 최대 / max = {hi}")
```

## 7.9 집합 자료형 / Set Data Type

**집합(set)** 은 순서가 없고 중복을 허용하지 않는 자료형입니다.  
A **set** is an **unordered** collection that does **not allow duplicates**.

- 문법 / Syntax: `{v1, v2, v3}` 또는 `set([...])`
- 빈 집합은 `set()` — `{}`는 빈 **딕셔너리**임 / empty set is `set()`, `{}` is an empty **dict**
- 순서가 없으므로 **인덱싱 불가** / no indexing (no order)
- 교집합, 합집합, 차집합 연산 지원 / supports union, intersection, difference

```python
# 집합 생성 / Creating sets
s1 = {1, 2, 3, 3, 2, 1}          # 중복은 자동으로 제거됨 / duplicates dropped
s2 = set([1, 2, 2, 3, 4])        # 리스트/튜플 → 집합 / from a list/tuple
s3 = set('hello')                 # 문자열 → 문자 집합 / from a string
empty = set()                     # 빈 집합 / empty set (NOT {})

print("s1:", s1)
print("s2:", s2)
print("s3:", s3)
print("empty:", empty, type(empty))
print("{} 의 타입 / type of {} :", type({}))   # 주의: {}는 딕셔너리 / {} is a dict!
```

```python
# 집합은 순서가 없으므로 인덱싱이 불가능 → TypeError
# Sets are unordered; indexing is not supported → TypeError
s = {10, 20, 30}
try:
    print(s[0])
except TypeError as e:
    print(f"TypeError 발생 / TypeError raised: {e}")
```

## 7.10 집합의 메소드 / Set Methods

| 연산 / Operation | 기호 / Operator | 메소드 / Method |
|---|---|---|
| 합집합 / Union | `\|` | `.union()` |
| 교집합 / Intersection | `&` | `.intersection()` |
| 차집합 / Difference | `-` | `.difference()` |
| 대칭 차집합 / Symmetric difference | `^` | `.symmetric_difference()` |

대칭 차집합: `(s1 ∪ s2) − (s1 ∩ s2)` — 양쪽에 공통이 아닌 원소들의 모임.  
Symmetric difference: elements in **either** set but **not both**.

```python
# 두 개의 집합을 정의 / define two sets
s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 6, 7, 8}

print("s1:", s1)
print("s2:", s2)

# 합집합 / union: | 또는 .union()
print("s1 | s2           =", s1 | s2)
print("s1.union(s2)      =", s1.union(s2))

# 교집합 / intersection: & 또는 .intersection()
print("s1 & s2              =", s1 & s2)
print("s1.intersection(s2)  =", s1.intersection(s2))

# 차집합 / difference: - 또는 .difference()
# s1에는 있지만 s2에는 없는 원소 / elements in s1 but not in s2
print("s1 - s2            =", s1 - s2)
print("s1.difference(s2)  =", s1.difference(s2))

# 대칭 차집합 / symmetric difference: ^ 또는 .symmetric_difference()
# 양쪽에 공통이 아닌 원소들 / elements in either, but not both
print("s1 ^ s2                       =", s1 ^ s2)
print("s1.symmetric_difference(s2)   =", s1.symmetric_difference(s2))
```

```python
# 여러 집합에 대한 연속 연산 / Chained operations on multiple sets
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
C = {4, 6, 7, 8}

# 세 집합의 합집합과 교집합
# Union and intersection of three sets
print("A | B | C =", A | B | C)
print("A & B & C =", A & B & C)

# 집합에 원소 추가/삭제 / add, remove elements
s = {1, 2, 3}
s.add(4)
s.add(2)          # 이미 있으므로 무시됨 / duplicate ignored
print("add 후 / after add:", s)
s.remove(1)       # 원소가 없으면 KeyError / KeyError if missing
s.discard(99)     # 원소가 없어도 오류 없음 / silent if missing
print("remove/discard 후 / after remove & discard:", s)
```

## 7.11 곱집합 / Cartesian Product (심화학습 / Advanced)

집합 A와 B의 **곱집합(Cartesian product)** $A \times B$는 $(a, b)$ 쌍의 집합입니다 (단, $a \in A$, $b \in B$).  
The Cartesian product $A \times B$ is the set of ordered pairs $(a, b)$ with $a \in A$ and $b \in B$.

$A^3 = A \times A \times A$ — 원소가 3개인 튜플들의 집합.

```python
# 슬라이드에 나온 product_set() 함수를 직접 구현
# Implement product_set() as in the slides
def product_set(a_set, b_set):
    """두 집합의 곱집합을 튜플의 집합으로 반환 / return Cartesian product as a set of tuples."""
    result = set()
    for a in a_set:
        for b in b_set:
            result.add((a, b))     # (a, b) 튜플을 집합에 추가 / add pair
    return result

A = {1, 2}
B = {'x', 'y', 'z'}

AxB = product_set(A, B)
print("A x B =", AxB)
print("원소 수 / size =", len(AxB), " ( = |A| * |B| =", len(A)*len(B), ")")
```

```python
# A^3 = A x A x A — product_set을 반복 적용하여 구현
# Apply product_set repeatedly for A^3
def product_power(a_set, n):
    """A^n 을 반환 / return A^n as a set of n-tuples."""
    # 초깃값: 1-튜플 집합 / start with 1-tuples
    result = {(x,) for x in a_set}
    for _ in range(n - 1):
        result = {tup + (x,) for tup in result for x in a_set}
    return result

A = {1, 2}
print("A^2 =", product_power(A, 2))
print("A^3 =", product_power(A, 3))
print("|A^3| =", len(product_power(A, 3)), " ( = |A|^3 = 8 )")
```

## 7.12 zip() 함수를 이용한 집적화와 슬라이싱 / Aggregation with zip() (심화학습 / Advanced)

- 리스트·딕셔너리·집합·튜플처럼 반복 가능한 자료형을 **반복가능(iterable) 자료형**이라 합니다.  
  Lists, dicts, sets, tuples, and strings are **iterables**.
- `zip()` 함수는 여러 반복가능 자료형을 **하나의 튜플 반복자**로 **집적(aggregation)** 합니다.  
  `zip()` aggregates multiple iterables into a single iterator of tuples.
- 인자가 $n$개이면 결과의 각 튜플은 $n$개의 원소를 가집니다.  
  With $n$ iterables, each resulting tuple has $n$ elements.

```python
# 인자가 없을 때: 공백 반복자 반환
# No arguments → empty iterator
print("zip() -> list:", list(zip()))   # []

# 인자가 하나일 때: 1-튜플들의 반복자
# One iterable → iterator of 1-tuples
single = list(zip(['a', 'b', 'c']))
print("zip(['a','b','c']) -> list:", single)
# 주의: 원소가 'a'가 아닌 ('a',) 임에 유의
# Note: the element is ('a',) (a 1-tuple), NOT just 'a'
```

```python
# 여러 개의 반복가능 자료형을 전달한 경우
# Multiple iterables as arguments
names  = ['Alice', 'Bob', 'Charlie']
ages   = [24, 30, 28]
jobs   = ['engineer', 'teacher', 'artist']

# 각 i번째 원소들이 하나의 튜플로 묶임
# The i-th elements are grouped into one tuple
aggregated = list(zip(names, ages, jobs))
print("집적화 / aggregated:")
for row in aggregated:
    print("  ", row)

# 길이가 다른 경우 → 가장 짧은 길이에 맞춰 잘림
# If lengths differ, zip() stops at the shortest
print("짧은 쪽에 맞춰짐 / truncated to shortest:",
      list(zip([1, 2, 3, 4], ['a', 'b'])))
```

```python
# 언집적화(unzip): zip된 자료를 원래 자료로 되돌리기
# 다른 함수가 아니라 zip()에 * 연산자를 붙여서 언패킹
# "Unzipping": pass the zipped data back to zip() with the * operator
pairs = [(1, 'a'), (2, 'b'), (3, 'c')]
nums, chars = zip(*pairs)     # * 연산자로 언패킹 / unpack with *
print("nums :", nums)          # (1, 2, 3)
print("chars:", chars)         # ('a', 'b', 'c')
```

## 7.13 튜플과 문자열의 슬라이싱 / Slicing Tuples and Strings

- 슬라이싱은 리스트뿐만 아니라 **튜플**과 **문자열**에도 동일하게 적용됨.  
  Slicing applies to **tuples** and **strings** just like lists.
- 문법 / Syntax: `seq[start:stop:step]`
- 음수 인덱스와 음수 스텝을 사용하면 **역순**으로 가져올 수 있음.  
  Negative indexes and negative steps traverse in reverse.

```python
# 튜플 슬라이싱 / Tuple slicing
t = (10, 20, 30, 40, 50, 60, 70)
print("t[1:4]   =", t[1:4])       # 인덱스 1,2,3  → (20, 30, 40)
print("t[:3]    =", t[:3])        # 처음 3개       / first 3
print("t[4:]    =", t[4:])        # 인덱스 4 이후  / from index 4
print("t[::2]   =", t[::2])       # 2칸씩         / every 2nd item
print("t[::-1]  =", t[::-1])      # 역순 / reverse
print("t[-3:-1] =", t[-3:-1])     # 음수 인덱스 / negative indexes
```

```python
# 문자열 슬라이싱 / String slicing
s = "Hello, Python!"
print("s[0:5]  =", s[0:5])        # 'Hello'
print("s[7:13] =", s[7:13])       # 'Python'
print("s[::-1] =", s[::-1])       # 문자열 뒤집기 / reverse the string
print("s[::2]  =", s[::2])        # 한 글자 건너뛰기 / every 2nd character
```
