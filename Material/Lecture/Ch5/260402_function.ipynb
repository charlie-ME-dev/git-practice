# Ch.5 함수와 입출력

```python
def print_star(n):
    for _ in range(n):
        print("*" * 10)

print_star(1)
print()
print_star(3)
print()
print_star(5)
```

```python
def draw_shape(symbol, width, height):
    for _ in range(height):
        print(symbol * width)

draw_shape("[ ]",8,3)
print()
draw_shape("ㅁ",5,4)
```
